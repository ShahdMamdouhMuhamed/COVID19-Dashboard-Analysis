# **Convert Date & Create Clean Copy**

```python
df_raw['date'] = pd.to_datetime(df_raw['date'])
df_clean = df_raw.copy()
print("Date column converted to datetime")
```

```python
df_raw.info()
```

# **Remove duplicates**

```python
rows_before = df_raw.shape[0]
print(f"Rows before removing duplicates: {rows_before:,}")

duplicate_count = df_raw.duplicated(subset=['location', 'date']).sum()
df = df_raw.drop_duplicates(subset=['location', 'date'])

rows_after = df.shape[0]
print(f"Rows after removing duplicates: {rows_after:,}")
```

# **Handle Missing Values**

```python
df_clean = df.copy()

print("Missing values BEFORE cleaning:", df_clean.isnull().sum().sum())

# New metrics → fill with 0
new_cols = [col for col in df_clean.columns if col.startswith('new_')]
df_clean[new_cols] = df_clean[new_cols].fillna(0)

# Cumulative totals → forward fill per country
cum_cols = ['total_cases', 'total_deaths', 'total_tests',
            'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
            'icu_patients', 'hosp_patients']
df_clean[cum_cols] = df_clean.groupby('location')[cum_cols].ffill()

# Constant columns (per country) → forward + backward fill
const_cols = ['continent', 'population', 'population_density', 'gdp_per_capita',
              'median_age', 'aged_65_older', 'life_expectancy', 'human_development_index',
              'extreme_poverty', 'diabetes_prevalence']
df_clean[const_cols] = df_clean.groupby('location')[const_cols].ffill().bfill()

# Remaining numeric columns → fill with 0
numeric_cols = df_clean.select_dtypes(include='number').columns
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(0)

print("Missing values AFTER cleaning :", df_clean.isnull().sum().sum())
```

```python
remaining = df_clean.isnull().sum()
remaining = remaining[remaining > 0].sort_values(ascending=False)
print("Columns that STILL have missing values (these will be dropped later):")
print(remaining)
```

# **Feature Engineering**

```python
import numpy as np
df_clean['case_fatality_rate'] = (df_clean['total_deaths'] / df_clean['total_cases'] * 100).replace([np.inf, -np.inf], 0).fillna(0)
df_clean['vaccination_rate'] = df_clean['people_fully_vaccinated_per_hundred']
df_clean['year'] = df_clean['date'].dt.year
df_clean['month'] = df_clean['date'].dt.month

print("added derived columns: case_fatality_rate, vaccination_rate, year, month")
```

# **Select Relevant Columns**

```python
selected_cols = [
    'iso_code', 'continent', 'location', 'date',
    'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
    'total_cases_per_million', 'new_cases_per_million',
    'total_deaths_per_million', 'new_deaths_per_million',
    'total_tests', 'new_tests', 'positive_rate', 'tests_per_case',
    'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
    'new_vaccinations_smoothed', 'total_vaccinations_per_hundred',
    'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred',
    'icu_patients', 'hosp_patients',
    'population', 'population_density', 'gdp_per_capita',
    'median_age', 'aged_65_older', 'life_expectancy', 'human_development_index',
    'extreme_poverty', 'diabetes_prevalence' , 'case_fatality_rate', 'vaccination_rate' , 'year', 'month'
]

df_clean = df_clean[selected_cols]
print(f"Reduced to {df_clean.shape[1]} relevant columns")
```

# **Handle outliers**

```python
print("Outlier detection (IQR method) - No removal performed")

key_cols = ['new_cases', 'new_deaths', 'total_cases_per_million', 'new_cases_per_million']

for col in key_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df_clean[col] < Q1 - 1.5*IQR) | (df_clean[col] > Q3 + 1.5*IQR)).sum()
    print(f"{col:25} → {outliers:,} potential outliers (kept as real events)")
```

# **Aggregate data**

```python
# ex: monthly summary per continent (for dashboard use)
monthly_agg = df_clean.groupby(['continent', 'year', 'month']).agg({
    'new_cases': 'sum',
    'new_deaths': 'sum',
    'total_vaccinations': 'max'
}).reset_index()

print("monthly aggregation created)")
print(monthly_agg.head())
```

# **Final Checks**

```python
print("Final Validation:")
print(f"Rows: {df_clean.shape[0]:,}")
print(f"Columns: {df_clean.shape[1]}")
print(f"Missing values: {df_clean.isnull().sum().sum()}")
print(f"Duplicate (location+date): {df_clean.duplicated(subset=['location','date']).sum()}")
print(f"Date range: {df_clean['date'].min().date()} → {df_clean['date'].max().date()}")
print("Data is clean and ready for dashboard")

df_clean.to_csv('clean_data.csv', index=False)
files.download('clean_data.csv')
df_clean.head()
```
